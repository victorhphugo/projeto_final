```python
import pandas as pd
url = "https://gist.githubusercontent.com/netj/8836201/raw/6f9306ad21398ea43cba4f7d537619d0e07d5ae3/iris.csv"
df = pd.read_csv(url)
df
```

```python
# troca automatica do caracter especial de todos os nomes de  colunas usando --str.replace('.','_')
df.columns = [s.replace('.','_')for s in df.columns]
df
```

```python
# traduzindo os nomes das colunas.
df.columns=['comprimento_sepala', 'largura_sepala', 'comprimento_petala', 'largura_petala', 'variedade']
df
```

```python
# verificando os tipos de plantas descritas na coluna "variedade" -- unique() ultilizado para ignorar os nomes duplicados.
df_variedade = sorted(df['variedade'].unique())
df_variedade
```

```python
# verificando quantas informações eu tenho de cada elemento da coluna "variedade"
df["variedade"].value_counts()
```

# observação: Estao bem diustribuidas

```python
# com base na visualização do grafico e no seu objetivo ,foi escolhido que é necessario visualizar o df com base nos nomes da plantas (na coluna  variedade para encontrar padroes por plantas)
df_index_variedade = df.set_index(["variedade"])
df_index_variedade
```

```python
df_index_variedade.describe().round(2)
```

```python
# guardando e separando por variaveis as especies das plantas na coluna.index("variedade")
df_info_setosa = df.query('variedade=="Setosa"')
df_info_versicolor = df.query('variedade=="Versicolor"')
df_info_virginica = df.query('variedade=="Virginica"')
```

```python
df_info_setosa.describe().round(2)
```
